```python
from textblob import TextBlob

# Test with sample sentences
sample_text = "This product is absolutely amazing! I love it so much."
blob = TextBlob(sample_text)

print("Text:", sample_text)
print("Polarity:", blob.sentiment.polarity)  # -1 (negative) to +1 (positive)
print("Subjectivity:", blob.sentiment.subjectivity)
```

```python
import pandas as pd

reviews = [
    "This product is amazing! Best purchase ever.",
    "Terrible quality, broke after one day. Very disappointed.",
    "It's okay, nothing special but works fine.",
    "Absolutely love it! Highly recommend to everyone.",
    "Worst experience ever. Complete waste of money.",
    "Decent product for the price, does the job.",
    "Outstanding quality and fast delivery!",
    "Not satisfied, customer service was unhelpful.",
    "Pretty good overall, met my expectations.",
    "Excellent! Exceeded all my expectations."
]

# Analyze sentiment for each review
results = []
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    results.append({'Review': review, 'Polarity': polarity, 'Sentiment': sentiment})

df = pd.DataFrame(results)
print(df)
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%',
        colors=['#90EE90', '#FFB6C1', '#FFFFE0'])
plt.title('Sentiment Distribution of Customer Reviews')
plt.show()

print("\nSentiment Summary:")
print(sentiment_counts)
```

```python
df.to_csv('sentiment_results.csv', index=False)
print("Saved to sentiment_results.csv")
```
